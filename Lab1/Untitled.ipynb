```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'Gas Turbine Dataset'
file_name = Path(path).glob('*.csv')
gt_df = pd.concat((pd.read_csv(file) for file in file_name), ignore_index=True)
```

```python
print(f"dataframe shape: {gt_df.shape}")
print(f"dataframe size: {gt_df.size}")
```

```python
print(f"Head:\n{gt_df.head()}")
```

```python
print(f"info:\n{gt_df.info()}")
```

```python
print(f"statistical description:\n{gt_df.describe()}")
```

```python
print(f"mode:\n{gt_df.mode()}")
```

```python
gt_df.hist(bins=100)
```

```python
cor_mat = gt_df.corr()
cor_mat_df = pd.DataFrame(cor_mat['TEY'].sort_values(ascending = False))
print(cor_mat_df)
```

```python
att = ['CDP', 'GTEP', 'TIT', 'AFDP']
gt_df_sub = gt_df[att]
print(gt_df_sub.head())
sns.pairplot(gt_df_sub)
```
